```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Schwarzschild radius
M = 1  # Mass of the black hole (in geometric units)
r_s = 2 * M  # Schwarzschild radius

# Define the radial range (r > r_s to avoid the singularity)
r = np.linspace(r_s * 1.1, 10 * M, 500)
theta = np.linspace(0, 2 * np.pi, 500)
r, theta = np.meshgrid(r, theta)

# Embedding function for the Schwarzschild equatorial plane
def schwarzschild_embedding(r, r_s):
    return np.sqrt(r_s * (r - r_s))

# Calculate the embedding surface
z = schwarzschild_embedding(r, r_s)

# Convert to Cartesian coordinates for plotting
x = r * np.cos(theta)
y = r * np.sin(theta)

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none', alpha=0.8)

# Plot Schwarzschild radius
ax.plot_surface(x, y, -z, cmap='viridis', edgecolor='none', alpha=0.8)
ax.set_title("Embedding Diagram of Schwarzschild Black Hole", fontsize=14)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

# Customize view angle
ax.view_init(elev=20, azim=45)

# Add a color bar
fig.colorbar(surf, shrink=0.5, aspect=10, label='Embedding Height')

plt.show()
```
